```python
import pandas as pd
```

```python
#url = 'https://docs.google.com/spreadsheets/d/17QXSa3sQ5JfF4uZVokEqL8eQbg5Pa9XN/edit#gid=1896038057'
#dataset = pd.read_csv('url')


iplDF = pd.ExcelFile('All_IPL_Match_Scores_easy.xlsx')
df = iplDF.parse('All_IPL_Data')
print(df)
```

```python
a = df.groupby('Year').Match_Number.count()
a
```

```python
b = df[df.Winning_Team == 'FirstBatting']
b
```

```python
c = b.groupby('Year').Match_Number.count()
c
```

```python
d = df[df.Winning_Team == 'Chasing'].groupby('Year').Match_Number.count()
d
```

```python
e = df[df.Winning_Team == 'Match Tied'].groupby('Year').Match_Number.count()
e
```

```python
ans = pd.concat([a,c,d,e],axis=1)
ans.fillna(0)
ans.columns = ['Total Number of Matches','Number of matches Team Batting First Won','Number of matches team Chasing Won','Number of times Match Tied']
ans
```
